#5. **HANDLING NA'S**

```python
# Handiling missing values (Dealing with NA)
#Find for NA's
olym.isna() # to find for NA's in a dataframe
```

```python
#Fill NA's with zero in dataset
olym.fillna(0) 
```

```python
#Fill NA's with previous value in data set
olym.fillna(method='pad')
```

```python
#Fill NA's with next values in dataset

olym.fillna(method ='bfill')
```

```python
#Fill NA's with a constant/text 
olym.fillna("Not available",inplace=True) # Note: this will modify any other views on this object
olym
```

```python
#Use interpolate function to fill the missing values which are in data set
olym.interpolate(method="linear") #â€˜linearâ€™: ignore the index and treat the values as equally spaced. default
```

```python
#Remove rows with NA's in dataset
olym.dropna() # Drop rows with NA
```

```python
#Remove row which contains all NA's

olym.dropna(how="all") # Drop the rows containing all NA's
```

```python
#Remove the columns containing NA's
olym.dropna(axis=1) # drop columns with NA's
```
